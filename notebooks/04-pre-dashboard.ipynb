### Importo librerías y el dataframe de mi EDA

```python
import pandas as pd
```

```python
df_bi = pd.read_csv('../data/output/ecommerce-for-dashboard.csv')
```

#### Elimino algunas variables para una ejecucción más eficiente en Power BI y enfocar mi dashboard al objetivo del análisis

```python
#defino las columnas y luego las elimino
cols_to_drop = [
    'customer_id', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 
    'order_estimated_delivery_date', 'customer_unique_id', 'order_item_id', 'price', 'review_id', 
    'review_comment_message', 'review_creation_date', 'effective_sale'
]

df_clean = df_bi.drop(columns=[c for c in cols_to_drop if c in df_bi.columns])
```

```python
#exporto el dataframe para mi visualización con PowerBI
df_clean.to_csv('../data/output/eda-for-dashboard.csv', index=False)
```

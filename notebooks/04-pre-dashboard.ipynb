### Importo librerías y el dataframe de mi EDA

```python
import pandas as pd
```

```python
df_bi = pd.read_csv('../data/output/ecommerce-for-dashboard.csv')
```

### Realizo algunos cálculos y agregaciones para enriquecer el dashboard en Power BI

```python
#creo una columna de agregación de estados por regiones
regiones_map = {
    'AM': 'Norte', 'RR': 'Norte', 'AP': 'Norte', 'PA': 'Norte', 'TO': 'Norte', 'RO': 'Norte', 'AC': 'Norte',
    'MA': 'Noreste', 'PI': 'Noreste', 'CE': 'Noreste', 'RN': 'Noreste', 'PE': 'Noreste', 'PB': 'Noreste', 'SE': 'Noreste', 'AL': 'Noreste', 'BA': 'Noreste',
    'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'DF': 'Centro-Oeste',
    'SP': 'Sudeste', 'RJ': 'Sudeste', 'MG': 'Sudeste', 'ES': 'Sudeste',
    'PR': 'Sur', 'RS': 'Sur', 'SC': 'Sur'
}

df_bi['region'] = df_bi['customer_state'].map(regiones_map)
```

```python
#creo una columna con el nombre completo de cada estado para facilitar el uso de mapas en Power BI
nombres_estados = {
    'AC': 'Acre', 'AL': 'Alagoas', 'AP': 'Amapa', 'AM': 'Amazonas', 'BA': 'Bahia',
    'CE': 'Ceara', 'DF': 'Distrito Federal', 'ES': 'Espirito Santo', 'GO': 'Goias',
    'MA': 'Maranhao', 'MT': 'Mato Grosso', 'MS': 'Mato Grosso do Sul', 'MG': 'Minas Gerais',
    'PA': 'Para', 'PB': 'Paraiba', 'PR': 'Parana', 'PE': 'Pernambuco', 'PI': 'Piaui',
    'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte', 'RS': 'Rio Grande do Sul',
    'RO': 'Rondonia', 'RR': 'Roraima', 'SC': 'Santa Catarina', 'SP': 'Sao Paulo',
    'SE': 'Sergipe', 'TO': 'Tocantins'
}

df_bi['customer_state_full'] = df_bi['customer_state'].map(nombres_estados)
```

#### Creo la variable *"delivery_dias"* que utilizaré para promedios de entregas en mi dashboard (en lugar de *"delta_customer"* y *"delivery_delta"* utilizadas para el análisis logístico)

```python
#aseguro que las variables tengan el formato fecha adecuado
df_bi['order_delivered_customer_date'] = pd.to_datetime(df_bi['order_delivered_customer_date'], errors='coerce')
df_bi['order_approved_at'] = pd.to_datetime(df_bi['order_approved_at'], errors='coerce')

#creo nueva columna que calcule el número de días entre la aprobación de la orden y la entrega del producto
df_bi['delivery_dias'] = (df_bi['order_delivered_customer_date'] - df_bi['order_approved_at']).dt.days
```

#### Elimino algunas variables para una ejecucción más limpia y eficiente en Power BI, y luego exporto el Dataset

```python
#defino las columnas y luego las elimino
cols_to_drop = [
    'customer_id', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 
    'order_estimated_delivery_date', 'shipping_limit_date', 'delivery_delta', 'delta_customer', 'review_id', 
    'review_comment_message', 'review_creation_date', 'effective_sale', 'valor_total', 'seller_delay', 'price'
]

df_clean = df_bi.drop(columns=[c for c in cols_to_drop if c in df_bi.columns])
```

```python
#exporto el dataframe para mi visualización con PowerBI
df_clean.to_csv('../data/output/eda-for-power-bi.csv', index=False)
```
